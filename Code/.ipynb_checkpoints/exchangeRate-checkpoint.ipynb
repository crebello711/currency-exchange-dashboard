```python
#import library
import pandas as pd
from IPython.display import Markdown, display

#read data
rawData = pd.read_excel("..\Data\Exchange_Rate_Report.xlsx")



#convert row 3 to header
rawData.columns = rawData.iloc[1]

'''drop row till October 30 1998,
    drop first two rows
    drop rows from index no 7340
'''
rawData.drop(rawData.index[[0,1]],inplace=True)
rawData.reset_index(drop=True,inplace=True)
rawData.drop(rawData.tail(18).index,inplace = True)
#Convert date time to dates
rawData['Date'] = pd.to_datetime(rawData['Date']).dt.date


#df_filtered = df[df['Age'] >= 25]
rawData =rawData.loc[rawData['Date'] > pd.to_datetime('1998-10-29')]
rawData.reset_index(drop=True,inplace=True)
display(rawData)

#export excel 
rawData.to_excel(r"C:\Users\cnr2231\git\Project_3\currency-exchange-dashboard\rawData.xlsx",index=True,header=True)
```
